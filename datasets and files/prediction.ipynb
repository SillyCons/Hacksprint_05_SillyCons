<h3>Linear Regression</h3>

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = data1[['2. high','3. low','1. open','5. volume']].values
Y = data1[['4. close']].values
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size=0.15,random_state=0)
LR1= LinearRegression()
LR1.fit(X_train,y_train)
LR1.score(X_test,y_test)
print(LR1.coef_)
print(LR1.intercept_)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
pred_lin= LR1.predict(X_test)
# R2 score evaluation on linear reg 
r2_lin = r2_score(y_test , pred_lin)
print("R2 score for LR: ", r2_lin)

#  Mean absoulte error on linear regression 
mae_lin = mean_absolute_error(y_test ,pred_lin)
print("Mean Absolute Error on LR: ",mae_lin)

# Mean Squared error on linear 
mse_lin = mean_squared_error(y_test , pred_lin)
print("Mean Squared Error on LR: ", mse_lin)
```

```python
df = {'Actual values':y_test ,'Predicted values':pred_lin}
df['Differences'] = df['Predicted values'] - df['Actual values']
df['Profit/Lost %'] = (df['Differences']*100)/ df['Actual values']
pd.DataFrame([df])
df
```

```python
df = data1[['4. close']]
df.head(5)
```

```python
future_days = 10
df['Predictions'] = df[['4. close']].shift(-future_days)
df.tail(5)
```

```python
X = np.array(df.drop(['Predictions'],1))[:-future_days]
print(X)
```

```python
# target dataset
y = np.array(df['Predictions'])[:-future_days]
print(y)
```

```python
# Split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.15)
```

```python
# models
# decision tree reg model
tree = DecisionTreeRegressor().fit(x_train,y_train)
# linear model
lr = LinearRegression().fit(x_train, y_train)
```

```python
x_future = df.drop(['Predictions'],1)[:-future_days]
x_future = x_future.tail(future_days)
x_future = np.array(x_future)
x_future
```

```python
# model tree pred
tree_prediction = tree.predict(x_future)
print(tree_prediction)
print()
# linear
lr_pred = lr.predict(x_future)
print(lr_pred)
```

```python
# visualize
predictions = tree_prediction
valid = df[X.shape[0]:]
valid['Predictions'] = predictions
plt.figure(figsize = (18,8))
plt.xlabel('Days')
plt.ylabel('Closing price')
plt.plot(df['4. close'])
plt.plot(valid[['4. close', 'Predictions']], color = 'r')
plt.legend(['Orig','Valid','Pred'])
plt.show()
```
